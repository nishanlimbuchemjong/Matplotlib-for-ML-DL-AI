```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('data/likes_ratio.csv')
data.head()
```

```python
view_count = data['view_count']
likes = data['likes']
ratio = data['ratio']
```

```python
plt.scatter(view_count, likes, c=ratio, cmap='summer', edgecolor='black', linewidth=1, alpha=0.75)

cbar = plt.colorbar()
cbar.set_label('Likes/Dislikes Ratio')
plt.xscale('log')
plt.yscale('log')
plt.tight_layout()
plt.title('Trending YouTube channel')
plt.xlabel('View Count')
plt.ylabel('Total Likes')
plt.show()
```
