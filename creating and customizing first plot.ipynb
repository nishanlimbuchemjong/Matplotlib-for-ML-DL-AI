```python
import matplotlib.pyplot as plt
```

# Plotting using raw datas

```python
x = [25, 26, 27, 29, 31, 34, 35, 37, 39, 40, 46]
y = [254, 263, 276, 297, 318, 348, 352, 376, 395, 403, 466]
```

```python
plt.plot(x, y)
plt.show()
```

```python
# Giving some titiles and labels to make it more informative
plt.plot(x, y)
plt.title('Median Salary by Age')
plt.xlabel('Ages')
plt.ylabel('Median Salary')
plt.show()
```
