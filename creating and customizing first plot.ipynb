```python
import matplotlib.pyplot as plt
```

# Plotting using raw datas

```python
# raw data of frontend developer
fx = [25, 26, 27, 29, 31, 34, 35, 37, 39, 40, 46]
fy = [254, 263, 276, 297, 318, 348, 352, 376, 395, 403, 466]
```

```python
# Giving some titiles and labels to make it more informative
plt.plot(fx, fy, label='Front-end Developer')

# raw data of backend developer
bx = [25, 26, 27, 29, 31, 34, 35, 37, 39, 40, 46]
by = [254, 263, 276, 297, 318, 348, 352, 376, 395, 403, 466]
plt.plot(bx, by, label='Back-end Developer')
plt.title('Median Salary by Age')
plt.xlabel('Ages')
plt.ylabel('Median Salary')
plt.legend()
plt.show()
```
