# Using Format String (color, linestyle)

```python
# Giving some titiles and labels to make it more informative
plt.plot(fx, fy, 'k--', label='Front-end Developer')  # 'k--'-> format string that says black color, dashed line

# raw data of backend developer
by = [232, 246, 257, 268, 312, 334, 354, 367, 374, 468, 486]
plt.plot(bx, by, 'g', label='Back-end Developer')  # 'g' --> green color and other is default
plt.title('Median Salary by Age')
plt.xlabel('Ages')
plt.ylabel('Median Salary')
plt.legend()
plt.show()
```

# Using Format String (color, linestyle, markers)
