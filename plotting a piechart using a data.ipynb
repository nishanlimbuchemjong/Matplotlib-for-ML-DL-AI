```python
import pandas as pd
import matplotlib.pyplot as plt
from collections import Counter
```

```python
data = pd.read_csv('data/data.csv')
```

```python
data.head()
```

```python
language_data = data['LanguagesWorkedWith']
language_data.head()
```

# Getting all languages with total no. of users

```python
language_counter = Counter()

for row in language_data:
    language_counter.update(row.split(';'))

language_counter
```

# Getting the top 5 languages

```python
languages = []
popularity = []

for row in language_counter.most_common(5):
    languages.append(row[0])
    popularity.append(row[1])

languages
```

```python
popularity
```

# Create a Pichart

```python
languages  # contains top 5 languages
popularity  # contains the total no. of users

explode = [0, 0, 0, 0.1, 0]
plt.pie(popularity, labels=languages, explode=explode, wedgeprops= {'edgecolor':'white'}, shadow=True, autopct='%1.1f%%')

plt.title("Top 5 Programming Languages")
plt.tight_layout()
plt.show()
```
